```python
model = keras.models.load_model("problem2_model")
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
def custom_predict(path):
    #this function take the path of the image then plots it and predict its class
    img = load_img(path, target_size=(60,60))
    image = plt.imread(path)
    plt.imshow(image)
    plt.axis("off")
    plt.show()
    img = img_to_array(img)
    img = tf.keras.applications.efficientnet.preprocess_input(img)
    img = img.reshape(1,60,60,3)
    pre = model.predict(img)
    p = np.argmax(pre)
    sign = int(list(class_indices.items())[p][0])
    print("this sign is of class " , sign," with " ,max(pre)[p]*100,"%  probability", sep ='')
```
